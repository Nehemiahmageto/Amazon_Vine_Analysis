```python
import numpy as np
import pandas as pd
```

```python
#Read in the vine_table.csv file as a DataFrame
df = pd.read_csv (r'vine_table.csv')
df.head()
```

```python
#Filter the data and create a new DataFrame where the total_votes count is equal to or greater than 20
helpful_reviews_df = df[df['total_votes'] >= 20] 
helpful_reviews_df.head()
```

```python
#create a new DataFrame where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
helpful_votes_df = helpful_reviews_df[helpful_reviews_df['helpful_votes']/helpful_reviews_df['total_votes'] >= 0.50]
helpful_votes_df.head()
```

```python
#a new DataFrame that retrieves all the rows of the Vine program (paid), vine == 'Y'.
vine_program_df = helpful_votes_df[helpful_votes_df['vine']=='Y']
vine_program_df.head()
```

```python
#retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
not_in_vine_program_df = helpful_votes_df[helpful_votes_df['vine']=='N']
not_in_vine_program_df.head()
```

```python
#Determine;total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews
vine_program_paid = {
    "total_number_of_reviews" : len(vine_program_df.index),
    "five_star_reviews" : len(vine_program_df[vine_program_df['star_rating'] == 5]),
    "five_star_reviews_percentage" : len(vine_program_df[vine_program_df['star_rating'] == 5]) / len(vine_program_df.index)
}

vine_program_unpaid = {
    "total_number_of_reviews" : len(not_in_vine_program_df.index),
    "five_star_reviews" : len(not_in_vine_program_df[not_in_vine_program_df['star_rating'] == 5]),
    "five_star_reviews_percentage" : len(not_in_vine_program_df[not_in_vine_program_df['star_rating'] == 5]) / len(not_in_vine_program_df.index)
}

results = {
    "vine_program_paid" : vine_program_paid,
    "vine_program_unpaid": vine_program_unpaid
}

print(results)
```
